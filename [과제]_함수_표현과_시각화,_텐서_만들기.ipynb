```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```
def my_func1(x):
  a=14
  b=25
  return a*x + b
```

```
def my_func2(x):
  a=7
  b=9
  c=-12
  return a*x**2 + b*x + c
```

```
def my_func3(x):
  a=2
  b=0
  c=14
  d=-7
  return a*x**3 + b*x**2 + c*x + d
```

```
x = np.arange(-5, 5, .1)
y1 = my_func1(x)
y2 = my_func2(x)
y3 = my_func3(x)

plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear function')
plt.show()

plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()

plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
plt.show()
```
