```
import numpy as np
```

```
def sigmoid(X):
  return 1/(1 + np.exp(-X))
```

```
def network():

  network = {}

  #첫번째 레이어
  network['W1'] = np.array([[.1, .3, .5], [.2, .4, .6], [.3, .5, .7]])
  network['B1'] = np.array([1,1,1])

  #두번째 레이어
  network['W2'] = np.array([[.2, .4, .6],[.1, .3, .5],[.4, .6, .8]])
  network['B2'] = np.array([1,1,1])

  #세번째 레이어
  network['W3'] = np.array([[.1, .3, .2],[-.1, -.5, -.4], [.3, .5, .2]])
  network['B3'] = np.array([1,1,1])

  #네번째 레이어
  network['W4'] = np.array([[.1, .2],[.3, .5],[.4, .6]])
  network['B4'] = np.array([1,1])

  #다섯번째 레이어
  network['W5'] = np.array([[.2, .4],[.2, .1]])
  network['B5'] = np.array([1,1])



  return network
```

```
def forward(network, x):
  W1, W2, W3, W4, W5 = network['W1'], network['W2'], network['W3'], network['W4'], network['W5']
  B1, B2, B3, B4, B5 = network['B1'], network['B2'], network['B3'], network['B4'], network['B5']

  A1 = np.dot(x, W1) + B1
  Z1 = sigmoid(A1)
  
  A2 = np.dot(Z1, W2) + B2
  Z2 = sigmoid(A2)
  
  A3 = np.dot(Z2, W3) + B3
  Z3 = sigmoid(A3)
  
  A4 = np.dot(Z3, W4) + B4
  Z4 = sigmoid(A4)
  
  A5 = np.dot(Z4, W5) + B5
  y = sigmoid(A5)

  return y
```

```
net = network()
x = np.array([.3, 1.3, -2.2])
y = forward(net, x)
print(y)
```
